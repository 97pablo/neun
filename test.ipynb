```python
import matplotlib.pyplot as plt

def points_from_txt(filename):
    x = []
    y = []
    with open(filename, "r") as f:
        for line in f:
            parts = line.strip().split()
            if len(parts) == 2:
                x.append(float(parts[0]))
                y.append(float(parts[1]))
                
    return x, y

def plot(file, label):
    # Load the data
    x, y = points_from_txt(file)
    plt.plot(x, y, label=label)

    
```

```python
plot("build/example1.txt", "basic")
plot("build/example2.txt", "optimized")

plt.xlabel("Time")
plt.ylabel("Voltage")
plt.title("Neuron graph")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
